```python
df = pd.read_csv('/datasets/project_sql_result_07.csv')

print(df.head())

bad_weather = df.query("weather_conditions == 'Bad'")['duration_seconds']
good_weather = df.query("weather_conditions == 'Good'")['duration_seconds']

alpha = 0.05

t_stat, p_value = stats.ttest_ind(bad_weather, good_weather, equal_var=False)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("❌ Rechazamos la hipótesis nula: hay diferencia significativa en la duración promedio.")
else:
    print("✅ No se rechaza la hipótesis nula: no hay evidencia suficiente de diferencia significativa.")
```

El análisis no mostró una diferencia estadísticamente significativa en la duración promedio de los viajes entre sábados lluviosos y sábados con un clima agradable (p > 0.05) Por lo tanto, no hay evidencia suficiente para afirmar que el clima lluvioso influye en la duración promedio de los viajes desde el barrio Loop hasta el Aeropuerto Internacional O'Hare
